```python
def get_cat_info(col):
    print("number of missing values:")
    print(data_train[col].isna().sum())
    print("missing values percentage:")
    print(data_train[col].isna().sum() / len(data_train))
    print("unique values:")
    print(data_train[col].value_counts())
    print("percentage:")
    print(data_train[col].value_counts() / len(data_train))
    print("null value / target : ")
    print(len(data_train[(data_train[col].isna()) & (data_train['target'] == 1)]) / len(data_train[data_train[col].isna()]))
    print("not null value : ")
    print(len(data_train[(data_train[col].notna()) & (data_train['target'] == 1)]) / len(data_train[data_train[col].notna()]))
```

```python
def get_reg_info(col):
    print("number of missing values:")
    print(data_train[col].isna().sum())
    print("missing values percentage:")
    print(data_train[col].isna().sum() / len(data_train))
    print(f"mean: {data_train[col].mean()}")
    print(f"median: {data_train[col].median()}")
    print(f"mode: {data_train[col].mode()[0]}")
    print(f"std: {data_train[col].std()}")

    print("\nEffect on results" + "\n" + "null value / target : ")
    print(len(data_train[(data_train[col].isna()) & (data_train['target'] == 1)]) / len(data_train[data_train[col].isna()]))
    print("not null value : ")
    print(len(data_train[(data_train[col].notna()) & (data_train['target'] == 1)]) / len(data_train[data_train[col].notna()]))
    
    plt.figure()
    plt.hist(data_train[col].dropna(), bins=50)
    plt.xlabel('Value')
    plt.ylabel('Count')
    plt.title(col)
    plt.show()
```
