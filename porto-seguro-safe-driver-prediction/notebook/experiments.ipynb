```python
xgb_scores = xgb_model.predict_proba(X_val)[:, 1]
```

```python
similarity_scores = -similarity_model.score_samples(X_val)
similarity_scores = (similarity_scores - similarity_scores.min()) / (similarity_scores.max() - similarity_scores.min())
```
